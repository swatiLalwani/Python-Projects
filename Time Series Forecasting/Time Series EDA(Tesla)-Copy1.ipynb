```python
# importing the yfinance package
import yfinance as yf

# giving the start and end dates
startDate = '2022-08-01'
endDate = '2024-08-01'

# setting the ticker value
ticker = 'TSLA'

# downloading the data of the ticker value between
# the start and end dates
resultData = yf.download(ticker, startDate, endDate)

# printing the last 5 rows of the data
print(resultData.tail())

# Setting date as index
resultData["Date"] = resultData.index

# Giving column names
resultData = resultData[["Date", "Open", "High","Low", "Close", "Adj Close", "Volume"]]

# Resetting the index values
resultData.reset_index(drop=True, inplace=True)

# getting the first 5 rows of the data
print(resultData.head())
```

```python
##Theoretical Part
```

Basic idea about time series data

1) Upward trend-  the increasing trend, mean will increase and variance will be same or less.
2) stationary data- mean will be in same line, it will help us to do many things like forecasting. we convert seasonal to stationary for forecasting.
3) Downward trend- the going down trend
4) cyclic data - for e.g. stocks and it is difficult to forecast, sales we can forecast
5) mean average (Rolling window)


## Simple moving Average

```python
df_tesla['Open'].plot(figsize=(15,6))
```

```python
df_tesla['Open:20 days rolling']=df_tesla['Open'].rolling(window=20,min_periods=1).mean()
df_tesla['Open:30 days rolling']=df_tesla['Open'].rolling(window=30,min_periods=1).mean()
df_tesla['Open:50 days rolling']=df_tesla['Open'].rolling(window=50,min_periods=1).mean()
```

```python
df_tesla[['Open','Open:20 days rolling','Open:30 days rolling','Open:50 days rolling' ]].plot(xlim =['2022-08-01','2024-08-01'],figsize=(15,6))
```

### CUMULATIVE AVERAGE (EXPANDING)

```python
df_tesla['Open'].expanding().mean().plot(figsize = (10,5))
```

## EMA - Exponential moving average

```python
##EMA Tesla Shares
## Smoothning Factor = 0.1
df_tesla['EMA_0.1']=df_tesla['Open'].ewm(alpha=0.1, adjust = False).mean()  #alpha is a smoothning factor
df_tesla['EMA_0.3']=df_tesla['Open'].ewm(alpha=0.3, adjust = False).mean()
```

```python
df_tesla[['Open','EMA_0.1','EMA_0.3' ]].plot(xlim =['2022-08-01','2024-08-01'],figsize=(15,6))
```

## Exponential Weighted moving Average

```python
df_tesla['EMA_5days']=df_tesla['Open'].ewm(span=5, adjust = False).mean()   #Span is used for weighted moving average
```

```python
df_tesla[['Open','EMA_5days' ]].plot(figsize=(15,6))
```
